### Soru 8
YetGen Harikalar Diyarın'da;

Her 7 saniyede bir doğum gerçekleşiyor.

Her 13 saniyede bir ölüm meydana geliyor.

Her 45 saniyede bir beyin göçü alınıyor.

Önümüzdeki 5 yıl sonraki popülasyonu hesaplamak için bir program yazın. Şu anki YetGen Harikalar Diyarı nüfusunun 2.000.000 kişi, 1 yılda 365 gün olduğunu varsayın.

İpucu : Pythonda bölme işlemi için tam sayı bölme operatörünü (//) kullanabilirsiniz. Örneğin; 5//4=1 (1.25 değil), 10//4=2 (2.5 değil) 

**Örnek**
```
>>>input
2000000
>>>output
15900484
```


````python
nufus = 2000000

saniye = 5*365*24*60*60
dogum = saniye//7
olum = saniye//13
beyinGocu= saniye//45

degisenNufus = dogum+beyinGocu-olum
toplamNufus = degisenNufus+nufus
print(toplamNufus)




````
