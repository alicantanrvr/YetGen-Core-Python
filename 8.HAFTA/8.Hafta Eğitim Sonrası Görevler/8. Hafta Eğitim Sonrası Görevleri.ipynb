# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
import time
 
def vki_hesaplama(kilo,boy):   
    vki = kilo/(boy**2)
    return round(vki,1)

def deger_araligi(vki):  
    
    if vki < 18.5:
        return "zayıf"
    elif 18.5 < vki < 24.9:
        return"Normal kilolu"
    elif 25 < vki < 29.9:
        return"Fazla kilolu"
    elif 30 < vki < 39.9:
        return "Obez"
    else:
        return"İleri obez"

def cikti_olustur(isim,  yas, kilo, boy, vki, deger):
    kaytit_tarihi = time.ctime(time.time())

    t1 = f"Kullanıcı adı : {isim}\n"
    t2 = f"Kullanıcı yaş, kilo ve boy bilgisi : {yas}, {kilo}, {boy}\n"
    t3 = f"Kullanıcının vucüt kitle indexi : {vki}\n"
    t4 = f"Kullanıcın değer araliği : {deger}\n"
    t5 = f"Kayıt tarihi : {kaytit_tarihi}\n"

    output = t1 + t2 + t3 + t4 + t5
    return output

def main():

    giris_sayisi = int(input("kaç kez giriş yapacaksınız"))

    for i in range(giris_sayisi):
        print(f"kullanıcı kaydı : {i +1}")
        isim = str(input("isminizi giriniz"))
        soyisim = str(input("soyisminizi giriniz:"))
        yas = int(input("yaşınızı giriniz"))
        boy = float(input("boyunuzu giriniz(m)"))
        kilo = float(input("kilonuzu giriniz(kg)")) 
        vki = vki_hesaplama(kilo, boy)
        deger = deger_araligi(vki)

        output = cikti_olustur(isim, soyisim, yas, kilo, vki, deger)

        with open("vki.txt","w",encoding="utf-8") as f:
            f.write(output)
            f.write("-"*20)
            print("Kullanıcı kaydedildi.")
    print("veriler 'vki.txt' dosyasına kaydedildi.")
main()
```

**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import time

menu = {
    "elma" : 2.5,
    "portakal" : 4.0,
    "kivi" : 5.0,
    "karpuz" : 3.0,
    "kavun" : 2.0
}

def display_menu():

    print("----YETBAR MENU----")

    for i in list(menu.keys()):

        print(f"{i.upper():10} : {menu[i]} TL")
    
def tek_urun_fiyat_hesaplama(urun,  adet):

    assert adet >= 0, "Ürün adedi 0 ve 0'dan küçük olamaz." #assert python dilinde bir hata ayıklama aracıdır.

    if urun not in list(menu.keys()):
        return None
    else:
        return menu[urun]* adet

def satin_alma():
    
    display_menu()
    print("Lütfen satın almak istedğiniz ürünün ismini giriniz. Çıkış yapmak içinse '0'ı basınız.")

    satin_alinanlar = []
    satilan_adet = []

    son_fiyat = 0

    while 1:
        eklenecek_urun = input("hangi ürünü eklemek istiyorsunuz?")

        if eklenecek_urun == "0":

            print("Fiş hazırlanıyor")
            break

        ekleme_adedi = int(input("kaç adet ürün eklemek istiyorsunuz?"))

        fiyat = tek_urun_fiyat_hesaplama(eklenecek_urun, ekleme_adedi)

        if fiyat is None:
            print("Yanlış bir işelm yaptınız, ürün eklenemedi.")
        else:
            print(f"Fiyat : {fiyat}.\n Satın alma işlemini onaylamak için'y' iptal etmek için'n' tuşlarına basınız.")
            check = input("Satın alma işlemini onaylamak istiyor musunuz?")

            if check == "y":
                print("İşlem onaynlandı.")
                print("-"*40)

                son_fiyat += fiyat
                satin_alinanlar.append(eklenecek_urun)
                satilan_adet.append(ekleme_adedi)

                print(f"Toplam Fiyat :{son_fiyat}")

            elif check == "n":

                print("Satın alma işlemi reddedildi.")
                print("-"*40)

    return son_fiyat ,satin_alinanlar, satilan_adet

def main():
    print("YETBAR'A HOŞGELDİNİZ!")
    print("Menüyü görünütlemek için '1'e, işlem yapmak için '2'ye basınız")

    program = None

    while 1 :
        program = int(input("Yapmak istediğiniz işlem : "))

        if program < 0 or program > 2:
            print("Yanlış bir işelm yaptınız. Tekrar deneyiniz")
        elif program == 0: 
            print("çıkış yapılıyor")
            break
        elif program == 1:
            display_menu()
        elif program == 2:
            satin_alma_ucreti, satin_alinanlar, satilan_adet  = satin_alma()

            with open("saglik.txt", "a", encoding="utf-8") as f:
                f.write(str(time.ctime(time.time())))
                f.write("\n")
                f.write(','.join(map(str, satin_alma())))
                f.write("\n")
                f.write(f"Toplam : {satin_alma_ucreti}")
                f.write("\n")
                f.write("-"*20)
                f.write("\n")

            kontrol = input("başka bir işlem yapmak istiyor musunuz?(evet-hayır)")

            if kontrol == "evet":
                pass
            else:
                print("TEŞEKKÜRLER")
                break

main()



```
