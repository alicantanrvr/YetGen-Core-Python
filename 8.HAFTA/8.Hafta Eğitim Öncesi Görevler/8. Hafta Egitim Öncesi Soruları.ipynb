# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
ad = str(input("ürünün ismini giriniz"))
fiyat = int(input("ürünün fiyatını giriniz"))
def urunKaydet():
    with open("urunler.txt", "a", encoding="utf-8") as f:
        f.write(f"{ad} : {fiyat}\n")
urunKaydet()
````

````python
with open('urunler.txt', 'r',encoding="utf-8") as f:
    print(f.read())
````

````python
def kelime_degistirme(yeni_kelime):
    
    with open('urunler.txt', 'r+',encoding="utf-8") as file:
        icerik = file.read()
        icerik = icerik.replace(ad, yeni_kelime)
        file.write(icerik)
kelime_degistirme("pantolon")

````

# Problem 3

```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```
