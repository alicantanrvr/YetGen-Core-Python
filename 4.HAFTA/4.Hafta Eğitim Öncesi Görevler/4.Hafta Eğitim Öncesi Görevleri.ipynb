# Problem 3

3 öğrenciden oluşan bir öğrenci not sözlüğü oluşturun. Bu sözlükte öğrencilerin notları value olarak bir listede toplansın.

Kullanıcıya hangi öğrencinin notlarını görmek istediğini sorun. Öğrencinin notu görüntülendiğinde program sonunda şöyle bir çıktı elde etmelisiniz:

```
Lütfen notlarını görmek istediğiniz öğrencinin adını girin: Mehmet

Mehmet isimli öğrencinin      1.Sınav Notu:72
                              2.Sınav Notu:66
                              3.Sınav Notu:48
Not Ortalaması: 62.0


````python
notlar = {
"Mehmet" : [85, 45, 64],
"Asya" : [75, 69, 82], 
"Ada" : [75, 96,28]
}

ogrenci = input("notlarını görüntülemek istediğiniz öğrencinin ismini giriniz: Mehmet")

if ogrenci in notlar:
    print(f"{ogrenci}, isimli öğrencinin birinci notu: {notlar[ogrenci][0]}")
    print(f"{ogrenci}, isimli öğrencinin ikinci notu: {notlar[ogrenci][1]}")
    print(f"{ogrenci}, isimli öğrencinin üçüncü notu: {notlar[ogrenci][2]}")

    not_ortlamasi = (notlar[ogrenci][0] +  notlar[ogrenci][1] +  notlar[ogrenci][2])/3
    print((f"Not ortalaması : {not_ortlamasi}"))

else: 
    print("yaptığınız işlemler yanlış, lütfen kontrol ediniz")
````
