# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle():
    def __init__(self, number):
        self.number = number
    
class Bus(Vehicle):
    def __init__(self, number):
        super().__init__(number)

    def seating_capasity(self):
        print(f"The seating capacity of a bus is {self.number} passnegers")
bus = Bus("temple")

passenger = Bus(50)
passenger.seating_capasity()
````
